```python
from pyscal_rdf import StructureGraph
```

```python
s = StructureGraph()
```

```python
struct = s.create.lattice.l12(lattice_constant=3.47, element=['Ni', 'Al'], repetitions=(2,2,2))
```

```python
s.visualise(backend='graphviz')
```

```python
gb = s.create.defect.grain_boundary(axis=[0,0,1], 
                        sigma=5, 
                        gb_plane=[3, -1, 0],
                        element='Fe')
```

```python
s.archive('data_2', compress=False)
```
